##### Batsman Comparison

```python
balls=df.groupby(['batsman'])['ball'].count().reset_index()
balls
```

```python
runs=df.groupby(['batsman'])['batsman_runs'].sum().reset_index()#total runs made by a batsman
runs.columns=['batsman','runs']
runs

###runs=df.groupby(['batsman'])['batsman_runs'].agg(sum).reset_index()
```

```python
df.groupby(['batsman'])['batsman_runs'].agg(sum)
```

```python
balls=balls.merge(runs,left_on='batsman',right_on='batsman',how='outer')
```

```python
four=df[df['batsman_runs']==4]
```

```python
''' df.groupby('batsman')['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index()'''

runs_4=four.groupby('batsman')['batsman_runs'].count().reset_index()
runs_4.columns=['batsman','4s']#number of times a batsman hits 4
runs_4
```

```python
six=df.groupby('batsman')['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index()
six.columns=['batsman','6s']#number of 6s made by a batsman
six
```

```python
player=pd.concat([runs,balls.iloc[:,1],runs_4.iloc[:,1],six.iloc[:,1]],axis=1)#combine all dataframes and we use axis=1 to concatenate vertically
#From balls df you select all rows but only 1st column similarly from other df as well.
player
```

```python
player.fillna(0,inplace=True)
player
```

```python
player['strike_rate']=(player['runs']/player['ball'])*100
player.head()
```

```python
player.isna().sum()
```

```python
player['4s'].fillna(0,inplace=True)
```

```python
player.isnull().values.any()
```

```python
df.columns
```

```python
grp=df.groupby(['match_id','batsman','batting_team'])['batsman_runs'].sum().reset_index()
grp
```

```python
max=grp.groupby(['batsman'])['batsman_runs'].max().reset_index()#max runs of individual players
max.columns=['batsman','max_runs']
max
```

```python
player2=pd.concat([player,max.iloc[:,1:3]],axis=1)#to concatenate 2 df (i.e. player and max )vertically we use axis=1 and we need all rows and 2columns from max df
player2
```
